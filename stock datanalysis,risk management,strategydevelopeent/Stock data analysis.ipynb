```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stocks = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'HDFCBANK.NS', 'ICICIBANK.NS']
data = yf.download(stocks, start='2024-04-04', end='2024-10-04')['Adj Close']
```

```python
daily_pct_change = data.pct_change().dropna()
```

```python
plt.figure(figsize=(14, 7))
for stock in stocks:
    plt.plot(data[stock], label=stock)
```

```python
plt.title('Closing Prices for Nifty 50 Stocks (Last 6 Months)')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend(loc='upper left')  # legend must be called after plotting with correct labels
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
for stock in stocks:
    plt.plot(daily_pct_change[stock], label=stock) 
```

```python
plt.title('Daily Percentage Changes for Nifty 50 Stocks (Last 6 Months)')
plt.xlabel('Date')
plt.ylabel('Daily Percentage Change')
plt.legend(loc='upper left')  # Call legend() after defining all labels
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
for stock in stocks:
    plt.plot(data[stock], label=f'{stock} Closing Price')  # Add label for closing price
    plt.plot(data[stock].rolling(window=20).mean(), label=f'{stock} 20-Day MA')  # Add label for moving average

plt.title('20-Day Moving Average vs Closing Prices for Nifty 50 Stocks')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')  # Ensure legend is called after defining all labels
plt.grid(True)
plt.show()
```
